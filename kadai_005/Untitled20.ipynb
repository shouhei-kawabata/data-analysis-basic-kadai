```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 3, tight_layout=True)
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x = np.linspace(0, 10, 200)
y = np.sin(x)
axes[0, 1].plot(x, y)

data = np.random.normal(loc=20, scale=5, size=1000)
axes[0, 2].hist(data, bins=20)

x_scatter = np.random.rand(100) * 8
y_scatter = np.random.rand(100) * 8
axes[1, 0].scatter(x_scatter, y_scatter)

plt.show()
```
